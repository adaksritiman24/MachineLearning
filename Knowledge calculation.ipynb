```python
import pandas as pd
import numpy as np
```

```python
data =pd.read_csv('play_tennis.csv')
```

```python
data['outlook'].value_counts()
data['Net']=1
```

```python
data
```

```python
data['play'].replace('No',0,inplace = True)
data['play'].replace('Yes',1,inplace = True)
data
```

```python
data=data.groupby('outlook').sum()
data['ent']=-((data['play']/data['Net'])*np.log2(data['play']/data['Net'])+((data['Net']-data['play'])/data['Net'])*np.log2((data['Net']-data['play'])/data['Net']))
```

```python
data['Total']=14
data['Total']= (data['Net']/data['Total'])*data['ent']
data
```
