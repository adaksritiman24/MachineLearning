```python
ratings.rename(columns ={'User-ID':'user_id','Book-Rating':'score'}, inplace =True)
```

```python
x = ratings['user_id'].value_counts()>200
```

```python
padhaku =x[x].index.tolist()
```

```python
books = books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
```

```python
books.head()
```

```python
books_with_rating =ratings.merge(books, on ='ISBN')
```

```python
books_with_rating.head()
```

```python
a=books_with_rating['ISBN'].value_counts()>50
```

```python
famous =a[a].index.tolist()
```

```python
len(famous)
```

```python
books_with_rating =books_with_rating[books_with_rating['ISBN'].isin(famous)]
```

```python
books_with_rating.shape
```

```python
books_with_rating.drop_duplicates(subset =['user_id','Book-Title'], inplace =True)
```

```python
books_with_rating.shape
```

```python
pt = books_with_rating.pivot_table(columns = 'user_id',index ='Book-Title', values ='score')
```

```python
pt.shape
```

```python
pt.fillna(-1,inplace = True)
```

```python
from scipy.sparse import csr_matrix
matrix =csr_matrix(pt)
```

```python
matrix
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
NN =NearestNeighbors(algorithm = 'brute')
```

```python
NN.fit(matrix)
```

```python
pt.shape
```
